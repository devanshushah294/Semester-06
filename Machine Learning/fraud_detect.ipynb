```python
dict_a = {0:"fraud",1:"non-fraud"}
print(df.Class.value_counts().rename(index=dict_a))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.loc[:,df.columns != 'Class']
y = df['Class']
train_x,test_x,train_y,test_y = train_test_split(x,y,test_size = 1/3,random_state=42,stratify=y)
```

```python
answer_dict = {}
```

```python
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
logist_reg = LogisticRegression(max_iter=1000)
logist_reg.fit(train_x,train_y)
y_pred = logist_reg.predict(test_x)
acc_logreg = round(accuracy_score(test_y,y_pred)*100,2)
answer_dict["logistic_regression"] = acc_logreg
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
model.fit(train_x,train_y)
y_pred = model.predict(test_x)
acc_nb = round(accuracy_score(test_y,y_pred)*100,2)
print(acc_nb)
answer_dict["gaussianNB"] = acc_nb
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(train_x,train_y)
y_pred = model.predict(test_x)
acc_dtc = round(accuracy_score(test_y,y_pred)*100,2)
print(acc_dtc)
answer_dict["decision_tree"] = acc_dtc
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
model.fit(train_x,train_y)
y_pred = model.predict(test_x)
acc_rfc = round(accuracy_score(test_y,y_pred)*100,2)
print(acc_rfc)
answer_dict["random_forest"] = acc_rfc
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier()
model.fit(train_x,train_y)
y_pred = model.predict(test_x)
acc_knc = round(accuracy_score(test_y,y_pred)*100,2)
answer_dict["k_neighbors_classifier"] = acc_knc
```
